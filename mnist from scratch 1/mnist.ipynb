# 2. Define Architecture

## Starting with a basic linear layer

```python
def init_params(size, std=1.0): return (torch.randn(size)*std).requires_grad_()
```

10 output classes required 

```python
output_classes = 10
weights = init_params((28*28, output_classes))
```

```python
weights.shape
```

```python
bias = init_params(1)
```

```python
def linear1(xb): return xb@weights + bias
preds = linear1(train_x)
preds[:5].shape
```

```python
bs = 10
dl = DataLoader(dset, batch_size = bs, shuffle=True)
```

```python
valid_dl = DataLoader(valid_dset, batch_size=bs, shuffle = True)
```

```python
x,y = first(dl)
y
```

```python
idx=3
show_image(x[idx].view(28, 28))
print(y[idx])
```

```python
train_dl = list(dl)
```

```python
acts = linear1(x)
acts.shape
```

```python
len(acts[0]), acts[0].sum()
```

Use softmax to get activations to sum to 1

```python
loss_func = nn.CrossEntropyLoss()
```

```python
loss_func(acts, y)
```

```python
def calc_grad(xb, yb, model):
    preds = model(xb)
    loss = loss_func(preds, yb)
    loss.backward()
```

```python
calc_grad(x, y, linear1)
weights.grad.mean(),bias.grad
```

```python
weights.grad.zero_()
bias.grad.zero_();
```

```python
def train_epoch(model, lr, params):
    for xb, yb in dl:
        calc_grad(xb, yb, model)
        for p in params:
            p.data -= p.grad*lr
            p.grad.zero_()
```

Use `torch.argmax(preds, dim=1)` to batch predict, the `dim=1` is important to get a prediction on each sample rather than the whole tensor 

```python
torch.argmax(preds, dim=1)
```

the wrong way...

```python
torch.argmax(preds)
```

```python
def batch_accuracy(xb, yb):
    preds = xb.sigmoid()
    correct = torch.argmax(xb, dim=1) == yb
    return correct.float().mean()
```

```python
batch_accuracy(linear1(x), y)
```

```python
def validate_epoch(model):
    accs = [batch_accuracy(model(xb), yb) for xb,yb in valid_dl]
    return round(torch.stack(accs).mean().item(), 4)
```

```python
validate_epoch(linear1)
```

```python
lr = 0.001
params = weights, bias
train_epoch(linear1, lr, params)
validate_epoch(linear1)
```

```python
for i in range(20):
    train_epoch(linear1, lr, params)
    print(validate_epoch(linear1), end=' ')
```

## Simplify things using the fastai SGD class

```python
def train_epoch(model):
    for xb,yb in dl:
        calc_grad(xb, yb, model)
        opt.step()
        opt.zero_grad()
```

```python
def train_model(model, epochs):
    for i in range(epochs):
        train_epoch(model)
        print(validate_epoch(model), end=' ')
```

```python
linear_model = nn.Linear(28*28,10)
opt = SGD(linear_model.parameters(), lr)
train_model(linear_model, 20)
```

```python
dls = DataLoaders(dl, valid_dl)
```

```python
learn = Learner(dls, nn.Linear(28*28,10), opt_func=SGD,
                loss_func=loss_func, metrics=batch_accuracy)
```

```python
learn.fit(10, lr=lr)
```

# Add some nonlinearity

```python
simple_net = nn.Sequential(
    nn.Linear(28*28,60),
    nn.ReLU(),
    nn.Linear(60,30),
    nn.ReLU(),
    nn.Linear(30,10)
)
```

```python
learn = Learner(dls, simple_net, opt_func=SGD,
                loss_func=nn.CrossEntropyLoss(), metrics=batch_accuracy)
```

```python
learn.fit(200, 0.001)
```

```python
plt.plot(L(learn.recorder.values).itemgot(2));
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.show()
```

```python
valid_x, valid_y  = first(valid_dl)
show_image(valid_x[0].view(28,28))
pred = torch.argmax(simple_net(valid_x)[0])
print(f'pred: {pred}, truth: {valid_y[0]}')
torch.argmax(simple_net(valid_x)[0]) == valid_y[0]
```
